# Categorical variables into Quantitative variables

```python
# 1. Using replace
```

```python
dir(df['gender'])
```

```python
df1 = df
genders = df1['gender'].unique()
genders
```

```python
df1['gender'].replace(genders, ['f','m'], inplace=True)
# df1['gender'].replace(genders, [0,1], inplace=True)
df1
```

```python
# 2. Using get_dummies method
```

```python
race_dummies = pd.get_dummies(df['race/ethnicity'], dtype=int)
race_dummies
```

```python
merged = pd.concat([df, race_dummies], axis=1)
merged
```

```python
merged.drop(['race/ethnicity'], axis='columns')
```

```python
df1 = df
```
