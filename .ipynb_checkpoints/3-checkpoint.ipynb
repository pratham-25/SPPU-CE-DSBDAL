```python
import pandas as pd
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head(10)
```

# Scan for Inconsistencies

```python
df.isnull().sum()
```

```python
# Drop colums

df.drop(['math score'], axis=1)
```

```python
# Drop rows

df.dropna(axis=0)
```

```python
df.head(10)
```

```python
# replace missing values with mean, median, mode -> Imputation
df['math score'].fillna(df['math score'].mean(), inplace=True)
df['reading score'].fillna(df['reading score'].median(), inplace=True)
df['writing score'].fillna(df['writing score'].mode().iloc[0], inplace=True)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

Data Normalisation

```python
df.dtypes
```

```python
# Min max scaling on math score

maxx = df['math score'].max()
minn = df['math score'].min()

df_norm = [(x - minn)/(maxx - minn) for x in df['math score']]

# Adding normalised column to our dataset
df["math score norm"] = df_norm

df.drop(['math score'],axis=1,inplace=True)
```

```python
df.head(10)
```

```python
# Z score scaling on reading score

mean = df['reading score'].mean()
stdd = df['reading score'].std()

df_z = [(x-mean)/stdd for x in df['reading score']]

# Adding normalised column to our dataset
df['reading score norm'] = df_z

df.drop(['reading score'], axis=1, inplace=True)
```

```python
df.head(10)
```

```python
print(df['reading score norm'].min(), df['reading score norm'].max())
```

# Normalising dataset 

```python
# !pip install seaborn
import seaborn as sns

sns.kdeplot(df['writing score'])
print("Skewness: ", df['writing score'].skew())
```

```python
# Converting into normalised dataset using sqrt transformation

df['writing score'] = np.sqrt(df['writing score'])
```

```python
sns.kdeplot(df['writing score'])
print("Skewness: ", df['writing score'].skew())
```

# Data Transformation

```python
def grade(x):
    if x<40:
        return 'F'
    elif x>=40 and x<50:
        return 'E'
    elif x>=50 and x<60:
        return 'D'
    elif x>=60 and x<70:
        return 'C'
    elif x>=70 and x<80:
        return 'B'
    elif x>=80 and x<90:
        return 'A'
    else:
        return 'A+'
```

```python
df['grade']=(df["math score norm"]+df["reading score norm"]+df["writing score"])/3
```

```python
df['grade'].apply(grade)
```
