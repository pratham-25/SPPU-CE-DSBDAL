```python
# Converting quantitative to categorical (age -> categories)

# print(df['age'].min(), df['age'].max())

bin_labels = ['young', 'middle', 'old']
bins = [0, 21, 50, 80]

df['age_cat'] = pd.cut(df['age'], bins=bins,labels=bin_labels) 
# df['age_cat'] = pd.cut(df['age'], bins=3, labels=bin_labels)
```

```python
df.head(10)
```

```python
group1 = df.groupby('age_cat')
group1.describe()
```

```python
young = group1.get_group('young')
young.head(10)
```

```python
young.mean(numeric_only=True)
```

```python
# Group by using categorical variable

grouped_data = df.groupby('age_cat')['salary']
summary1 = grouped_data.describe()
summary1
```

```python
# Group by using aggregation of columns

grouped_data1 = df.groupby(['groups','age_cat'])['salary']
summary2 = grouped_data1.describe()
summary2
```
