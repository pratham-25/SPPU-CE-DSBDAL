# Grouping of Quantitative variables by Categorical Variables

```python
grouped_data = df.groupby('species')['sepal_length']
grouped_data.describe()
```

```python
grouped_data = df.groupby('species')['sepal_width']
grouped_data.describe()
```

```python
grouped_data = df.groupby('species')['petal_length']
grouped_data.describe()
```

```python
grouped_data = df.groupby('species')['petal_width']
grouped_data.describe()
```

```python
grouped_data = df.groupby('species')

for group_name, group_data in grouped_data:
    print(f"Species Name: {group_name}")
    print("-----------------------------")
    print(group_data.describe())
    print()
```
